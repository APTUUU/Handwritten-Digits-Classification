````python
import numpy as np
import tensorflow as tf
````

# Constants
<hr>

````python
X_TEST_PATH = 'MNIST/load_xtest.csv'
Y_TEST_PATH = 'MNIST/load_ytest.csv'
````

# Load the Data
<hr>

````python
y_test = np.loadtxt(Y_TEST_PATH, delimiter=',', dtype=int)
print(f'Shape of y_test {y_test.shape}')
````

````python
x_test = np.loadtxt(X_TEST_PATH, delimiter=',', dtype=float)
print(f'Shape of x_test {x_test.shape}')
````

# Create Session
<hr>

````python
graph = tf.compat.v1.Graph()

session = tf.compat.v1.Session(graph=graph)
````

# Load Model
<hr>

````python
tf.compat.v1.saved_model.load(session,
                           [tf.compat.v1.saved_model.tag_constants.SERVING],
                           'SavedModel-512-64 LR0.001 E18')
````

# Run Session & Make Prediction
<hr>

````python
X = graph.get_tensor_by_name('X:0')
Y_pred = graph.get_tensor_by_name('Accuracy_Calculation/ArgMax:0')
````

````python
prediction = session.run(fetches=Y_pred, feed_dict={X:x_test})
````

# Check Results
<hr>

````python
correct_pred = sum(np.equal(prediction, y_test))
total_pred = y_test.shape[0] 
````

````python
correct_prc = correct_pred / total_pred
print('Accuracy is {:.2%}'.format(correct_prc))
````

# Cleanup
<hr>

````python
session.close()
tf.compat.v1.reset_default_graph()
````
